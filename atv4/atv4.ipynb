```python
from qiskit import QuantumCircuit, execute, Aer, quantum_info
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
```

```python
#1.1
np.random.seed(seed=0)
n = 10000
alice_bits = np.random.randint(2, size=n)
alice_bases = np.random.randint(2, size=n)
key = []
for i in range(n):
    qc = QuantumCircuit(1,1)
    if alice_bases[i] == 0:
        if alice_bits[i] == 0:
            pass
        else:
            qc.x(0)
    else:
        if alice_bits[i] == 0:
            qc.h(0)
        else:
            qc.x(0)
            qc.h(0)
    qc.barrier()
    key.append(qc)
```

```python
#1.2
bob_bases = np.random.randint(2, size=n)
backend = Aer.get_backend('qasm_simulator')
measurements = []
for q in range(n):
    if bob_bases[q] == 0:
        key[q].measure(0,0)
    if bob_bases[q] == 1:
        key[q].h(0) # mudando base de medida
        key[q].measure(0,0)
    result = execute(key[q], backend, shots=1, memory=True).result()
    measured_bit = int(result.get_memory()[0])
    measurements.append(measured_bit)
bob_results = measurements
```

```python
#1.3
alice_key = []
bob_key = []
for i in range(n):
    if alice_bases[i] == bob_bases[i]:
        alice_key.append(alice_bits[i])
        bob_key.append(bob_results[i])
print("alice's key: "+str(alice_key)+"\n")
print("bob's key:   "+str(bob_key))
print(bob_key.count(0)+bob_key.count(1))
```

```python
#1.4
n_sample = int(n/10)
idx_qber = np.random.randint(n, size=n_sample)
alice_sample = []
bob_sample = []
for i in idx_qber:
    i = np.mod(i, len(alice_key))
    alice_sample.append(alice_key.pop(i))
    bob_sample.append(bob_key.pop(i))
qber = 0.0
for i in range(n_sample):
    if alice_sample[i] != bob_sample[i]:
        qber = qber+1
qber = qber/n_sample
print(qber)
```
